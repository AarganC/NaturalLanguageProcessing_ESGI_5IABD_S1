```python
import warnings 
import gensim 

from gensim.models import Word2Vec 
from nltk.tokenize import sent_tokenize, word_tokenize 
  
warnings.filterwarnings(action = 'ignore') 
```

```python
sample = open("../../files/cgu_figaro.txt")
s = sample.read()
```

```python
f = s.replace("\n", " ")
data = []
```

```python
# iterate through each sentence in the file 
for i in sent_tokenize(f): 
    temp = [] 
      
    # tokenize the sentence into words 
    for j in word_tokenize(i): 
        temp.append(j.lower()) 
  
    data.append(temp) 
```

```python
data
```

```python
# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count = 1,  
                              size = 100, window = 5) 
# Print results 
print("Cosine similarity between 'CGU' " + 
               "and 'site' - CBOW : ", 
    model1.similarity('cgu', 'site')) 
      
print("Cosine similarity between 'CGU' " +
                 "and 'mutuellement' - CBOW : ", 
      model1.similarity('cgu', 'mutuellement')) 
```
